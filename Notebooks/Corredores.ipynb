```python
import pandas as pd
```

```python
url = "https://raw.githubusercontent.com/AlexisG81/etl-data-pipeline/refs/heads/main/data/raw/corredores.csv"
```

```python
df = pd.read_csv(url)
```

```python
df.head()
```

```python
df.shape

df.columns

df.info()

df.isnull().sum()
```

```python
df.columns = df.columns.str.strip().str.lower()

for col in df.select_dtypes(include='object'):
    df[col] = df[col].str.strip()
```

```python
df['zona'] = df['zona'].fillna('No especificado')
```

```python
df['nivel'] = df['nivel'].fillna('Desconocido')
```

```python
df['anios_experiencia'] = df['anios_experiencia'].fillna(df['anios_experiencia'].mean())
```

```python
df['anios_experiencia'] = df['anios_experiencia'].astype(int)
```

```python
df = df.drop_duplicates()
```
