```python
import pandas as pd
```

```python
url="https://raw.githubusercontent.com/AlexisG81/etl-data-pipeline/refs/heads/main/data/raw/polizas.csv"
```

```python
df = pd.read_csv(url)
```

```python
df.head()
```

```python
df.shape

df.columns

df.info()

df.isnull().sum()
```

```python
df['fecha_emision'] = pd.to_datetime(df['fecha_emision'], errors='coerce')
```

```python
df['prima'] = df['prima'].astype(str).str.replace(",", ".")
df['comision'] = df['comision'].astype(str).str.replace(",", ".")
df['monto_asegurado'] = df['monto_asegurado'].astype(str).str.replace(",", ".")
```

```python
df['prima'] = pd.to_numeric(df['prima'], errors='coerce')
df['comision'] = pd.to_numeric(df['comision'], errors='coerce')
df['monto_asegurado'] = pd.to_numeric(df['monto_asegurado'], errors='coerce')
```

```python
df['prima'] = df['prima'].fillna(df['prima'].mean())
df['comision'] = df['comision'].fillna(df['comision'].mean())
df['monto_asegurado'] = df['monto_asegurado'].fillna(df['monto_asegurado'].mean())
```

```python
df = df.drop_duplicates()
```
